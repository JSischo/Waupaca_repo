```python
import pandas as pd
import csv
from sklearn.preprocessing import LabelEncoder
```

```python
# File to Load
medical_data_to_load = "Resources/medical_data.csv"
```

```python
# Read the medical data and store into a Pandas DataFrame
medical_data_df = pd.read_csv(medical_data_to_load)
```

```python
medical_data_df
```

## Animal type encoded:  cat=0, dog=1, rabbit=2

```python
#encoding animal type column into new df, clean_medical_df 

le = LabelEncoder()

clean_medical_df = medical_data_df.copy()
clean_medical_df['animal_type'] = le.fit_transform(clean_medical_df['animal_type'])
```

```python
# delete record_no and status, "record_type" non-beneficial columns

clean_medical_df = clean_medical_df.drop(columns=['record_no','status', 'record_type'])
```
