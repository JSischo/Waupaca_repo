```python
df.rename(columns = {"Intake Type": "Intake_Type"}, inplace=True)
df.rename(columns = {"Intake Subtype": "Intake_Subtype"}, inplace=True)
df
```

```python
clinic_address_df = df[df.Intake_Type == "Clinic"]
```

```python
clinic_address_df = df[(df.Intake_Subtype == "Spay") | (df.Intake_Subtype == "Neuter")]
```

```python
# create separate dataframe for out address with intake type = clinic, subtype spay or neuter, 
# process these dataframes separately from df
# remove nans before converting to lat/long (not important to keep index of df)

clinic_address_df = clinic_address_df[['Animal #', "Species", "Intake_Type","Intake_Subtype",'Out Street Address','Out City','Out Province','Out Postal Code']]
clinic_address_df
```

```python
# remove Nans from addresses dfs, see how many rows are left

clinic_address_df = clinic_address_df.dropna()

clinic_address_df
```

```python
# these are the datapoints that will be mapped to show where clinic customers are located, 
#columns need to be concat to run in geolocater to get lat/long
```

```python
clinic_address_df.rename(columns = {"Out Street Address": "Out_Street_Address"}, inplace=True)
clinic_address_df.rename(columns = {"Out City": "Out_City"}, inplace=True)
clinic_address_df.rename(columns = {"Out Province": "Out_Province"}, inplace=True)
clinic_address_df.rename(columns = {"Out Postal Code": "Out_Postal_Code"}, inplace=True)
clinic_address_df
```

```python
# Using separate cleaned address dataframes, concat separate address columns into one for lat/long conversion

clinic_address_df["clean_clinic_address"] = clinic_address_df['Out_Street_Address'].map(str) + '  ' + clinic_address_df['Out_City'].map(str) + '  ' + clinic_address_df['Out_Province'].map(str) + '  ' + clinic_address_df['Out_Postal_Code'].map(str)
clinic_address_df
```

```python
#this may be used on separate addresses dataframes
#EXAMPLE INPUT ADDRESS/OUTPUT using Nominatim

geolocator = Nominatim(timeout=10, user_agent = "myGeolocator")
#location = geolocator.geocode('4550 Kester Mill Rd,Winston-Salem,NC')
#print(location)
#print((location.latitude, location.longitude))
```

```python
#use GeoPy/Nominatim to convert Found_Address and Adopter_Address to lat/long
# add columns to store lat/long 

clinic_address_df['clinic_geocode'] = clinic_address_df.clean_clinic_address.apply(geolocator.geocode)
```

```python
clinic_address_df
```

## Address conversion to lat/long - removal of Nones

```python
# obtain lat/long for each address

clinic_address_df['clinic_lat'] = [g.latitude if g is not None else 0 for g in clinic_address_df.clinic_geocode]
clinic_address_df['clinic_long'] = [g.longitude if g is not None else 0 for g in clinic_address_df.clinic_geocode]
```

```python
clinic_address_df
```

```python
clinic_address_df = clinic_address_df[clinic_address_df.clinic_lat != 0]
clinic_address_df
```

```python
clinic_address_df = clinic_address_df.drop(columns=['Out_Street_Address', 'Out_City', 'Out_Province', 'Out_Postal_Code', 'clean_clinic_address', "clinic_geocode"])
clinic_address_df
```
